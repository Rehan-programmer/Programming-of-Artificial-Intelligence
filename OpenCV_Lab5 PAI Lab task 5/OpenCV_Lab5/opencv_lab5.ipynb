# OpenCV Basics (Lab 5)
This notebook demonstrates main OpenCV functions.

```
import cv2
import numpy as np
```

## Read and Display Image

```
img = cv2.imread('sample.jpg')
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Resize Image

```
resized = cv2.resize(img, (300,300))
cv2.imshow('Resized', resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Convert to Grayscale

```
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Draw Shapes

```
cv2.line(img, (0,0), (200,200), (255,0,0), 2)
cv2.rectangle(img, (50,50), (150,150), (0,255,0), 2)
cv2.circle(img, (100,100), 40, (0,0,255), 2)
cv2.putText(img, 'OpenCV', (50,250), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
cv2.imshow('Shapes', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Edge Detection

```
edges = cv2.Canny(gray, 100, 200)
cv2.imshow('Edges', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
